````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils
from imutils.object_detection import non_max_suppression
````

````python
im = plt.imread("/Users/madatbay/Downloads/cv/cat.jpg")
````

````python
plt.imshow(im)
````

### Create Cascase Classifier for cats

````python
cat = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml')
````

### Identify rectangle of detected objects

````python
cats = cat.detectMultiScale(cv2.cvtColor(im, cv2.COLOR_BGR2GRAY))
````

````python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in cats])
````

````python
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.8)
````

````python
for (x1, y1, x2, y2) in rects_NMS:
    cv2.rectangle(im, (x1, y1), (x2, y2), (255, 0, 0), 2)
````
